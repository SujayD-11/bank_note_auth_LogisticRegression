```python
#Libraries used are : Pandas, Numpy, Sci-Kit Learn, Matplotlib, Seaborn.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
import seaborn as sb
```

```python
df = pd.read_csv('BankNoteAuthentication.csv')
df.head()
```

```python
df.shape
```

1372 rows and 5 columns. 4 Input varibles and 1 output.

```python
df.describe
```

```python
features_with_nan = [features for features in df.columns if df[features].isnull().sum()>=1]

for features in features_with_nan:
    print(features,np.round(df[features].isnull().mean(),4), '% missing values')
```

Hence we can conclude that no nan values are present.

```python
features_with_num = [features for features in df.columns if df[features].dtypes != 'O']
print("Number of numerical variables: ",len(features_with_num))
```

All 5 features are Numerical in nature and last can be said as categorical but data is numerical. 1=not authentic while 0=authentic.

```python
features_with_num #list of features
```

```python
for feature in features_with_num:
    data  = df.copy()#good practice to copy the dataset into new var
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

Looking at the above histograms we can say that 4 input variables are continuos, while the output variable is discrete.
