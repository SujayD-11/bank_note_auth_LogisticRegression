```python
#Libraries used are : Pandas, Numpy, Sci-Kit Learn, Matplotlib, Seaborn.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
import math
import pickle

pd.pandas.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('new_Bank_Auth.csv')
df.head()
```

```python
X = df[['variance','skewness','curtosis']]
Y = df['class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
```

80-20 split

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
auth_pred = model.predict(x_test)
```

```python
accuracy_score(y_test,auth_pred)
```

```python
filename = 'linear_regression.pkl'
pickle.dump(model, open(filename, 'wb'))
```
